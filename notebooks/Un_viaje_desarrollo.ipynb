# Un viaje por el gasto turístico en Canarias

# INTRODUCCIÓN

Canarias es uno de los destinos turísticos más importantes de Europa, y podríamos decir que tiene un superpoder muy particular: el clima es una primavera eterna.

Mientras en buena parte del continente se pelea con la lluvia y el frío, en Canarias hay personas tomando el sol en pleno febrero. Para un turista europeo, eso no es solo atractivo: es casi milagroso.

Pero el encanto de Canarias no se limita al buen tiempo. Es un archipiélago con personalidad múltiple: ocho islas, ocho mundos, ocho formas diferentes de vivir unas vacaciones sin salir del mismo territorio. Y ahí está la clave del turismo canario: la diversidad.

#### Tenerife, por ejemplo, es la isla “todoterreno”.

Playas, montaña, ciudades históricas, parques temáticos, gastronomía, excursiones al Teide… Una isla que ofrece desde playas familiares hasta rutas volcánicas. Su turismo es variado, internacional y con ganas de gastar en ocio y experiencias.

### Gran Canaria es la más cosmopolita.
Aquí hay dunas, sí… pero también centros comerciales, museos, congresos, vida cultural y una capital que vive de cara al mar. El turista de Gran Canaria no solo busca hamaca: también busca tiendas, espectáculos, gastronomía y vida urbana.

### Lanzarote es otra historia:
Volcanes, silencio, paisajes negros y rojos, arquitectura integrada en el entorno y la huella irresistible de César Manrique. Es un destino elegante, tranquilo y artístico. Su visitante quiere descansar, comer bien y disfrutar de un paisaje único en el mundo.

### Luego está Fuerteventura, paraíso de viento y arena infinita.
Surf, kitesurf, playas kilométricas, dunas que parecen sacadas del desierto y un ritmo de vida lento. Su turista es fiel: quien la descubre, vuelve… y vuelve.

### La Palma, la isla verde, respira naturaleza pura:
Bosques, volcanes, senderos, miradores y un cielo tan limpio que permite ver galaxias enteras. Uno de los destinos más ligados al ecoturismo y a las experiencias al aire libre.

### La Gomera es la isla donde el tiempo se detiene.
Laurisilva, acantilados, silencio y rutas que parecen cuentos. El turismo aquí es calmado, respetuoso y orientado a desconectar de verdad.

### El Hierro, pequeña pero sorprendente:
Auténtica, tranquila, volcánica, sostenible, perfecta para el buceo y para el viajero que quiere tranquilidad absoluta. Es turismo de calidad, aunque en menor volumen.

### Y La Graciosa, la joya sin prisas:
sin asfalto, sin estrés, con bicicletas, arena y mar como protagonistas. Un lugar para desaparecer del ritmo del mundo durante unos días.

### ¿Qué significa todo esto?
Que Canarias no funciona como un solo destino, sino como un conjunto de modelos turísticos que conviven dentro del mismo archipiélago.
Y si las islas son diferentes, también lo es la forma en que los turistas gastan su dinero.

Por eso, más allá de saber cuántos turistas llegan, la verdadera pregunta es:

✅ ¿Cuánto gasta un turista cuando viene a Canarias?

✅ ¿En qué se lo gasta?

✅ ¿Gasta lo mismo un turista alemán que uno español o uno nórdico?

✅ ¿Y qué isla atrae a los turistas que más consumen?

Para intentar responder a estas cuestiones, se han formulado cuatro hipótesis sobre el comportamiento del gasto turístico en las islas:

### HIPÓTESIS

#### 1. Son Los turistas del norte de Europa son los que más gastan por día
Viajan desde países con mayor poder adquisitivo, se alojan en establecimientos más caros y consumen más en ocio, restauración y actividades, verdad o mito?

#### 2. La mayor parte del gasto turístico se concentra en el alojamiento.
Los hoteles, apartamentos y viviendas vacacionales representan el núcleo principal del presupuesto total del viaje.

#### 3. Tenerife es la isla donde más dinero se gasta.
Porque tiene más infraestructuras turísticas, más oferta de ocio, más actividades y mayor capacidad hotelera.

#### 4. Tras la pandemia, el gasto medio por turista se ha recuperado e incluso aumentado.
Menos viajes al año, pero viajes más largos, con más comodidad y más disposición a gastar.

En resumen, Canarias no solo es un destino turístico; es una economía viva donde cada visitante deja una huella económica distinta. Entender cómo gastan los turistas es comprender cómo se mueve el motor económico de las islas. Y con estas hipótesis damos el primer paso para analizarlo.


En concreto, trabajamos con:

   - Gasto medio por turista y día

   - Desglose del gasto por conceptos
     (alojamiento, restauración, transporte, ocio, compras, etc.)

   - Series por islas, para comparar territorios

   - Series por nacionalidades, para identificar diferencias entre países de origen

   - Series temporales por años y trimestres, desde 2018 hasta la actualidad

Estos datos permiten conocer no solo cuánto gastan los turistas, sino dónde, cuándo y cómo reparten ese gasto, y por eso son fundamentales para estudiar el impacto económico del turismo en Canarias.

Para asegurar la comparabilidad entre períodos, se incorporan estimaciones para el tercer y cuarto trimestre de 2025. Estas estimaciones se calculan utilizando el promedio de los valores registrados en los terceros y cuartos trimestres de los años anteriores. 

Se emplean los datos del periodo 2018–2024, puntualizar que `los cálculos para el análisis de las hipótesis planteadas se realizan a nivel anual` y no trimestre a trimestre, con el fin de mantener la comparabilidad entre períodos analizados. 

Esta metodología permite obtener una aproximación coherente con el comportamiento histórico del gasto en Canarias sin necesidad de aplicar modelos estadísticos complejos, como medias ponderadas o regresiones.

""" Calculo la media de los terceros y cuartos trimestres de las anualidades en estudio, para estimar el gasto en cada una de las partidas en estudio (para cada fila del df).
Hago una lista llamada cols_3T, en la que itero por cada una de las columnas para que en cada columna que encuentre el 3T o el 4T, me saque la cifra de esa columna, para luego poder hacer el promedio de trimestre indicado. Además le indico que redondee a dos decimales.
 Creo un bucle para comprobar, itero por las columnas para ver si ya existe la el nombre de la columna, y si está ya que la elimine para que no haya duplicidades """


```python
# Detectar automáticamente las columnas que contienen "3T" o "4T"
cols_3T = [col for col in df.columns if "3T" in col]
cols_4T = [col for col in df.columns if "4T" in col]


# Calcular los promedios y redondear
prom_3T = df[cols_3T].mean(axis=1).round(2)
prom_4T = df[cols_4T].mean(axis=1).round(2)

# Eliminar columnas si ya existen para evitar errores
for col in ["2025 3T", "2025 4T"]: 
    if col in df.columns:
        df.drop(columns=[col], inplace=True)

# Insertar en la cuarta y quinta columna (después del índice)
df.insert(4, "2025 3T", prom_3T)  # posición 3 → cuarta columna
df.insert(3, "2025 4T", prom_4T)  # posición 4 → quinta columna

# Revisar
df.head()
```

Con el fin de efectuar los cálculos necesarios, se incorpora una nueva columna, 2025, que registra para cada fila el promedio anual de los cuatro trimestres, considerando los dos trimestres extraídos de los datos y los dos restantes estimados.

""" Necesito crear una nueva columna, 2025, con la media de los 4 trimestres, lo hago con df.iloc[:, 3:7] → selecciona las columnas 3, 4, 5 y 6, por posición(Python cuenta desde 0 y el límite superior no se incluye).Calculo la media con .mean, como antes y redondeo a dos decimales.
Como me ha dado fallo, le indico que si existe una columna que se llame 2025, que la elimine

Con df.insert(3, '2025', prom_2025) le indico que quiero que la columna nueva me la ponga en la cuarta posición del df."""


```python
# Eliminar columnas antiguas que puedan existir
for col in ["2025", "Promedio 2025"]:
    if col in df.columns:
        df.drop(columns=[col], inplace=True)

# Calcular promedio de las columnas que quieras (3,4,5,6)
prom_2025 = df.iloc[:, 3:7].mean(axis=1).round(2)

# Insertar la nueva columna en la posición deseada (cuarta columna)
df.insert(3, "2025", prom_2025)

# Revisar resultado
df.head()
```

Para facilitar la interpretación de la Hipótesis 2, los datos de gasto se han expresado en porcentajes. Esto permite comparar de manera clara la proporción que representa cada tipo de gasto en el total anual y visualizar tendencias a lo largo de los años.

```python
# Se extraen los datos de todos los años en estudio
cols_anios = ['2025','2024','2023','2022','2021','2020','2019','2018']
# para calcular los porcentajes de cada tipo de gasto sobre el total
df_porcentaje_col = df_comprobacion_h2.copy()
df_porcentaje_col[cols_anios] = df_porcentaje_col[cols_anios].div(
    df_porcentaje_col[cols_anios].sum(axis=0), axis=1
) * 100
# para que redondee a dos decimales
df_porcentaje_col[cols_anios] = df_porcentaje_col[cols_anios].round(2)

df_porcentaje_col
```

```python
# Convertir a formato long
df_long = df_porcentaje_col.melt(
    id_vars=['gasto'], 
    value_vars=cols_anios, 
    var_name='Año', 
    value_name='Porcentaje'
)

# Gráfico de barras
plt.figure(figsize=(12,6))
sns.barplot(x='Año', y='Porcentaje', hue='gasto', data=df_long)

# Indicamos las etiquetas de los ejes y el tipo de leyenda

plt.title('Distribución porcentual del gasto en Canarias por tipo')
plt.ylabel('Porcentaje (%)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python

# Definir la paleta personalizada por isla
paleta_islas = {
    'Tenerife': '#6baed6',        # azul intenso
    'Fuerteventura': '#FFEA00',   # amarillo
    'Gran Canaria': '#ff7f0e',    # naranja
    'Lanzarote': '#a6cee3',       # gris-azulado suave
    'La Palma': '#b2df8a'         # verde suave
}

# Configuración general de Seaborn
sns.set_style("whitegrid")
plt.figure(figsize=(12,6))

# Crear gráfico de barras
ax = sns.barplot(
    data=df_max_por_anio,
    x='Año',
    y='Valor máximo',
    hue='Isla',
    palette=paleta_islas
)

# Añadir títulos y etiquetas
plt.title('Valor máximo por año y isla', fontsize=16)
plt.ylabel('Valor máximo')
plt.xlabel('Año')

# Añadir valores encima de cada barra
# Añadir valores encima de cada barra, más grandes y en negrita
for p in ax.patches:
    height = p.get_height()
    ax.annotate(
        f'{height:.2f}',           
        (p.get_x() + p.get_width() / 2., height),
        ha='center', va='bottom',
        fontsize=12,        # tamaño más grande
        fontweight='bold',  # negrita
        color='black',
        xytext=(0, 3),
        textcoords='offset points'
    )

# Leyenda fuera del gráfico
plt.legend(title='Isla', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
