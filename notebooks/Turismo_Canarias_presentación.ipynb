# Un viaje por el gasto turstico en Canarias

Canarias es uno de los destinos tursticos ms importantes de Europa, y podramos decir que tiene un superpoder muy particular: el clima es una primavera eterna.

Mientras en buena parte del continente se pelea con la lluvia y el fro, en Canarias hay personas tomando el sol en pleno febrero. Para un turista europeo, eso no es solo atractivo: es casi milagroso.

Pero el encanto de Canarias no se limita al buen tiempo. Es un archipilago con personalidad mltiple: ocho islas, ocho mundos, ocho formas diferentes de vivir unas vacaciones sin salir del mismo territorio. Y ah est la clave del turismo canario: la diversidad.


#### Tenerife, por ejemplo, es la isla todoterreno.

Playas, montaa, ciudades histricas, parques temticos, gastronoma, excursiones al Teide Una isla que ofrece desde playas familiares hasta rutas volcnicas. Su turismo es variado, internacional y con ganas de gastar en ocio y experiencias.

### Gran Canaria es la ms cosmopolita.
Aqu hay dunas, s pero tambin centros comerciales, museos, congresos, vida cultural y una capital que vive de cara al mar. El turista de Gran Canaria no solo busca hamaca: tambin busca tiendas, espectculos, gastronoma y vida urbana.

### Lanzarote es otra historia:
Volcanes, silencio, paisajes negros y rojos, arquitectura integrada en el entorno y la huella irresistible de Csar Manrique. Es un destino elegante, tranquilo y artstico. Su visitante quiere descansar, comer bien y disfrutar de un paisaje nico en el mundo.

### Luego est Fuerteventura, paraso de viento y arena infinita.
Surf, kitesurf, playas kilomtricas, dunas que parecen sacadas del desierto y un ritmo de vida lento. Su turista es fiel: quien la descubre, vuelve y vuelve.

### La Palma, la isla verde, respira naturaleza pura:
Bosques, volcanes, senderos, miradores y un cielo tan limpio que permite ver galaxias enteras. Uno de los destinos ms ligados al ecoturismo y a las experiencias al aire libre.

### La Gomera es la isla donde el tiempo se detiene.
Laurisilva, acantilados, silencio y rutas que parecen cuentos. El turismo aqu es calmado, respetuoso y orientado a desconectar de verdad.

### El Hierro, pequea pero sorprendente:
Autntica, tranquila, volcnica, sostenible, perfecta para el buceo y para el viajero que quiere tranquilidad absoluta. Es turismo de calidad, aunque en menor volumen.

### Y La Graciosa, la joya sin prisas:
sin asfalto, sin estrs, con bicicletas, arena y mar como protagonistas. Un lugar para desaparecer del ritmo del mundo durante unos das.

### Qu significa todo esto?
Que Canarias no funciona como un solo destino, sino como un conjunto de modelos tursticos que conviven dentro del mismo archipilago.
Y si las islas son diferentes, tambin lo es la forma en que los turistas gastan su dinero.


Por eso, ms all de saber cuntos turistas llegan, la verdadera pregunta es:

 Cunto gasta un turista cuando viene a Canarias?

 En qu se lo gasta?

 Gasta lo mismo un turista alemn que uno espaol o uno nrdico?

 Y qu isla atrae a los turistas que ms consumen?

Para intentar responder a estas cuestiones, he formulado cuatro hiptesis sobre el comportamiento del gasto turstico en las islas:

### HIPTESIS

#### 1. Son Los turistas del norte de Europa son los que ms gastan por da
Viajan desde pases con mayor poder adquisitivo, se alojan en establecimientos ms caros y consumen ms en ocio, restauracin y actividades, verdad o mito?

#### 2. La mayor parte del gasto turstico se concentra en el alojamiento.
Los hoteles, apartamentos y viviendas vacacionales representan el ncleo principal del presupuesto total del viaje.

#### 3. Tenerife es la isla donde ms dinero se gasta.
Porque tiene ms infraestructuras tursticas, ms oferta de ocio, ms actividades y mayor capacidad hotelera.

#### 4. Tras la pandemia, el gasto medio por turista se ha recuperado e incluso aumentado.
Menos viajes al ao, pero viajes ms largos, con ms comodidad y ms disposicin a gastar.


En resumen, Canarias no solo es un destino turstico; es una economa viva donde cada visitante deja una huella econmica distinta. Entender cmo gastan los turistas es comprender cmo se mueve el motor econmico de las islas. Y con estas hiptesis damos el primer paso para analizarlo.

Toda la informacin y los datos analizados proceden del Instituto Canario de Estadstica (ISTAC), que es el organismo oficial encargado de recopilar y publicar la informacin estadstica sobre el turismo en Canarias.

En concreto, trabajamos con:

   - Gasto medio por turista

   - Gasto medio por turista y da

   - Desglose del gasto por conceptos
     (alojamiento, restauracin, transporte, ocio, compras, etc.)

   - Series por islas, para comparar territorios

  - Series por nacionalidades, para identificar diferencias entre pases de origen

  - Series temporales por aos y trimestres, desde 2018 hasta la actualidad

Estos datos permiten conocer no solo cunto gastan los turistas, sino dnde, cundo y cmo reparten ese gasto, y por eso son fundamentales para estudiar el impacto econmico del turismo en Canarias.

```python
import pandas as pd

# Lista de archivos excel para convertir en csv
archivos_xlsx = [
    "gastomedioxestaciaxturista.xlsx",
    "gastoxturistaxdia.xlsx"
]

for archivo in archivos_xlsx:
    # Para que cargue el archivo Excel
    df = pd.read_excel(archivo)

    # Crear el nombre del CSV cambiando la extensin
    archivo_csv = archivo.replace(".xlsx", ".csv")

    # Guardar como CSV
    df.to_csv(archivo_csv, index=False)
```
